```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos todo el dataset
pc = pd.read_csv("../data/train.csv")

# Mostramos todas las filas 
# print("Todas las filas del dataset")
# display(pc)

# Mostramos solo las primeras filas del dataset
print("Las primeras filas del dataset")
display(pc.head())

# Cargamos y mostramos la info del dataset en general
print("Informacion general del dataset")
display(pc.info())

# Sacamos con describe todo lo general
print("Sacamos estadisticas numéricas generales")
display(pc.describe())

# Sacamos con describe una columna en concreto
print("Sacamos estadisticas numéricas")
display(pc["Age"].describe())

# Sacamos con describe una columna en concreto
print("Sacamos estadisticas numéricas")
display(pc["Pclass"].describe())

# Si queremos más de una columna, hacemos tal que esto
print("Sacamos estadisticas numéricas")
display(pc[["Age","Sex"]].describe())

# Con value_counts para ver cuantos supervivientes
print("Conteo de supervivientes")
display(pc["Survived"].value_counts())

# Con value_counts para ver cuantos iban en primera,segunda,tercera clase
print("Conteo de supervivientes")
display(pc["Pclass"].value_counts())

# Gráfico para ver cuantos hombres y mujeres murieron
sns.countplot(data=pc, x="Sex", hue="Survived")
plt.title("Supervivencia por sexo")
plt.show()

# Grafico para mostrar cuantas personas murieron 
sns.countplot(data=pc, x="Pclass", hue="Survived")
plt.title("Supervivencia por clase")
plt.show()

```
